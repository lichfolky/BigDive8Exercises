```python
df = df.dropna(subset=('duration', 'director_name'))
```

```python
dirLen = df.director_name.apply(lambda x : len(str(x)))
dirLen.head()
```

```python
dfSel = df[['duration', 'director_name']]
dfSel.head()
```

```python
res = dfSel.join(dirLen, rsuffix='N')
res.head()
```

```python
res.plot.scatter('duration', 'director_nameN')
```

```python
import numpy as np
import scipy as sp
np.corrcoef(res.duration, res.director_nameN)
```
